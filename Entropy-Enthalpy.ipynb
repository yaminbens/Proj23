```python
import numpy as np
import plumed
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8, 8))

for i,l in enumerate(["metadNa300", "metadNa350", "metadNa400"]):
    datarw = plumed.read_as_pandas("../../Vashti/"+l+"/colvar/COLVAR")

    kbt=350*1.380649*6.02/1000
    c_s,b_s = np.histogram(datarw["s2"][50000:],bins=100,density=True,weights=np.exp(datarw["metad.rbias"][50000:]/kbt))
    x_s = 0.5*(b_s[:-1]+b_s[1:])
    fes_s = -np.log(c_s) #-beta logP
    fes_s -= fes_s.min()
    c_h,b_h = np.histogram(datarw["enthalpyPerAtom"][50000:],bins=30,density=True,weights=np.exp(datarw["metad.rbias"][50000:]/kbt))
    x_h = 0.5*(b_h[:-1]+b_h[1:])
    fes_h = -np.log(c_h) #-beta logP
    fes_h -= fes_h.min()

    H, xedges2, yedges2 = np.histogram2d(datarw["enthalpyPerAtom"][50000:],datarw["s2"][50000:], 
                                         bins=100, density=True,weights=np.exp(datarw["metad.rbias"][50000:]/kbt))
    H = H.T
    fes2D2 = np.log(H)
#     fes2D2 -= fes2D2.min()
    
    ax1 = fig.add_subplot(3,3,i+1)
    ax1 = plt.plot(x_s,fes_s)
    # plt.ylim(0,20)
    plt.ylabel("FES")
    plt.xlabel("s2")
    plt.xlim(-7,-1)
    plt.title(l)
    ax2 = fig.add_subplot(3,3,i+4)
    ax2 = plt.plot(x_h,fes_h)
    plt.ylim(0,20)
    plt.xlim(-104,-94)
    plt.ylabel("FES")
    plt.xlabel("enthalpyPerAtom")
    plt.title(l)
    im1 = fig.add_subplot(3,3,i+7)
    fig.set_figheight(15)
    fig.set_figwidth(15)
    im1 = plt.imshow(fes2D2, interpolation='nearest', aspect='auto',
                     origin='lower',extent=[xedges2[0], xedges2[-1], yedges2[0], yedges2[-1]],vmin=-15)#,vmax=25)
    plt.xlabel("enthalpyPerAtom")
    plt.ylabel("s2")
    plt.ylim(-7,-1)
    plt.xlim(-104,-94)
    plt.title(l)
    plt.colorbar()
```

```python
dGs = []
for j,l in enumerate(["metadNa300", "metadNa350", "metadNa400"]): 
    datarw = plumed.read_as_pandas("../../Vashti/"+l+"/colvar/COLVAR")
    dG =[]
    for i in range(100):
        kbt=350*1.380649*6.02/1000

        H, xedges2, yedges2 = np.histogram2d(datarw["enthalpyPerAtom"][0:3000*(i+1)],
                                             datarw["s2"][0:3000*(i+1)], 
                                             bins=100, density=True,
                                             weights=np.exp(datarw["metad.bias"][0:3000*(i+1)]/kbt))


        H=H/np.sum(H)

        HS = H[(-102<=xedges2[:-1]) & (xedges2[:-1]<-99),:][:,(-4.5<=yedges2[:-1]) & (yedges2[:-1]<-3.25)]
        HL = H[(-98.5<xedges2[:-1]) & (xedges2[:-1]<=-96.5),:][:,(-3.25<yedges2[:-1]) & (yedges2[:-1]<=-2.25)]
        dG.append(-kbt*np.log(np.sum(HL)/np.sum(HS)))
    dGs.append(dG)



```

```python
plt.plot(np.arange(0,300,3), dGs[0], label = "300K") 
plt.plot(np.arange(0,300,3), dGs[1], label = "350K") 
plt.plot(np.arange(0,300,3), dGs[2], label = "400K") 


# labels
plt.xlabel("Simulation time")
plt.ylabel("DeltaG [kJ/mol]")
plt.legend()
plt.show()
```

```python
import math 

# calculate free-energy difference
DeltaF=[]; kbt=2.494339
# cycle on free-energy profiles
for i in range(0,321):   
 # import fes file into pandas dataset
 data=plumed.read_as_pandas("../../Vashti/metadNa350/fes/fes_"+str(i)+".dat")
 # find minimum value of fes
 minf = min(data["file.free"])
 # plot fes
 F0=0.0; F1=0.0
 for j in range(0, len(data["s2"])):
     s2 = data["s2"][j]
     # calculate probability
     p = math.exp((-data["file.free"][j]+minf)/kbt)
     # integrate in the two basins
     if(-3<=s2<=-2):   F0 += p
     if(-7<=s2<=-3.5): F1 += p
 # calculate free energy difference and add to list
 DeltaF.append(-kbt*math.log(F0/F1))         
# plot
plt.plot(DeltaF) 
# labels
plt.xlabel("Simulation time")
plt.ylabel("DeltaF [kJ/mol]")
```
